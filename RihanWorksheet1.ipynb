```python
import time

list1 = [i for i in range(1_000_000)]
list2 = [i for i in range(1_000_000)]

start = time.time()
result_list_add = [list1[i] + list2[i] for i in range(len(list1))]
end = time.time()
print("Python List Addition Time:", end - start, "seconds")
```

```python
import numpy as np
import time

arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)

start = time.time()
result_np_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", end - start, "seconds")
```

```python
start = time.time()
result_list_mul = [list1[i] * list2[i] for i in range(len(list1))]
end = time.time()
print("Python List Multiplication Time:", end - start, "seconds")

```

```python
start = time.time()
result_np_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", end - start, "seconds")


```

```python
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()
print("NumPy Array Dot Product Time:", end - start, "seconds")

```

```python
matrix1 = [[i for i in range(1000)] for _ in range(1000)]
matrix2 = [[i for i in range(1000)] for _ in range(1000)]

start = time.time()
result_matrix = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(1000)) for j in range(1000)] for i in range(1000)]
end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")
```
